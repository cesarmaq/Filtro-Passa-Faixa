```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

```python
f = np.logspace(2,6,2000)
```

```python
w = 2*np.pi*f
```

```python
L = 200e-6
```

```python
C = 51e-9
```

```python
R  = 47
```

```python
XC = 1./(w*C)
```

```python
XL = w*L
```

```python
Z = (cmath.sqrt(-1)*XL*cmath.sqrt(-1)*XC)/(cmath.sqrt(-1)*XL-cmath.sqrt(-1)*XC)
```

```python
HW = Z/(R+Z)
```

```python
# plot interactivity in notebook
%matplotlib notebook
```

```python
plt.figure()
plt.semilogx(f,abs(HW))
plt.grid()
plt.title('Resposta em FrequÃªncia do Filtro Passa Faixa (H(w))')
plt.xlabel('FrequÃªncia (Hz)')
plt.ylabel('Amplitude (V)')
plt.show()
```
